```python
import pandas as pd

# Load the datasets
day_df = pd.read_csv('C:/Users/Tanisha/Desktop/bike/day.csv')
hour_df = pd.read_csv('C:/Users/Tanisha/Desktop/bike/hour.csv')

# Display the first few rows of both datasets
print(day_df.head())
print(hour_df.head())
```

```python
# Check for missing values
print(day_df.isnull().sum())
print(hour_df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plots
sns.set(style="whitegrid")

# Plot daily total ridership
plt.figure(figsize=(10, 6))
sns.lineplot(data=day_df, x="dteday", y="cnt", label="Total Riders")
plt.xticks(rotation=45)
plt.title("Daily Total Ridership Over Time", fontsize=16)
plt.xlabel("Date", fontsize=20)
plt.ylabel("Total Riders", fontsize=20)
plt.show()
```

```python
# Plot hourly total ridership
plt.figure(figsize=(10, 6))
sns.lineplot(data=hour_df, x="hr", y="cnt", label="Total Riders")
plt.title("Hourly Total Ridership", fontsize=16)
plt.xlabel("Hour of the Day", fontsize=12)
plt.ylabel("Total Riders", fontsize=12)
plt.show()
```

```python
# Plot registered and casual users - Daily
plt.figure(figsize=(10, 6))
sns.lineplot(data=day_df, x="dteday", y="casual", label="Casual Riders")
sns.lineplot(data=day_df, x="dteday", y="registered", label="Registered Riders")
plt.xticks(rotation=45)
plt.title("Daily Registered vs Casual Riders", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Riders", fontsize=12)
plt.legend()
plt.show()
```

```python
# Analyzing weather influence on ridership
plt.figure(figsize=(10, 6))
sns.barplot(data=day_df, x="weathersit", y="cnt")
plt.title("Effect of Weather on Ridership", fontsize=16)
plt.xlabel("Weather Situation", fontsize=12)
plt.ylabel("Total Riders", fontsize=12)
plt.show()
```
